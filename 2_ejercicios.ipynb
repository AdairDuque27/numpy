```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # 1. Intentamos leer la imagen original
    img = plt.imread('beagle.jpg')

    # 2. Si se lee correctamente, aplicamos el efecto negativo
    img_negativo = 255 - img

    # 3. Mostramos el resultado en pantalla
    plt.imshow(img_negativo)
    plt.title("Imagen en Negativo")
    plt.show()

except FileNotFoundError:
    print("❌ ERROR: No se encontró el archivo 'beagle.jpg' en esta carpeta.")
    print("Asegúrate de que la foto del perrito esté guardada exactamente en el mismo lugar que este notebook.")
```

```python
# Copiamos la imagen original para no destruirla
img_roja = img.copy()

# Ponemos en 0 los canales 1 (Verde) y 2 (Azul)
img_roja[:, :, 1] = 0
img_roja[:, :, 2] = 0

plt.imshow(img_roja)
plt.title("Solo Canal Rojo")
plt.show()
```
